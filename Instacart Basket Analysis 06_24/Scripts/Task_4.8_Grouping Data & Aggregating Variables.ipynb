# 4. & 5. Creating column for average spender


```python
df_merged['average_spend'] = df_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_merged [['user_id' ,'order_number' ,'loyalty_flag', 'average_spend']].head()
```

 # 6. To check different types of spenders like Low spender & High spender.

```python
df_merged.loc[df_merged['average_spend'] < 10, 'spender_flag'] = 'Low_spender'
```

```python
df_merged.loc[df_merged['average_spend'] >= 10, 'spender_flag'] = 'High_spender'
```

```python
df_merged['spender_flag'].value_counts(dropna = False)
```

```python
df_merged [['user_id' ,'order_number' ,'loyalty_flag', 'average_spend', 'spender_flag', 'prices']].head(100)
```

# 7.To determine frequent versus non-frequent customers

```python
# creating column 
df_merged['customer_frequency'] = df_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

# Creating flag 

```python
df_merged.loc[df_merged['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_merged.loc[(df_merged['customer_frequency'] > 10 ) & (df_merged['customer_frequency'] <= 20) ,'loyalty_flag'] = 'Regular customer'
```

```python
df_merged.loc[df_merged['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_merged['customer_frequency'].value_counts(dropna = False)
```

```python
df_merged [['user_id' ,'order_number' ,'loyalty_flag', 'average_spend', 'spender_flag', 'prices','customer_frequency']].head(100)
```
